```python
#Importing ast module fro extracting information

import ast
```

```python
# 'genres' and 'keywords' field contain significant information in the 'name' key of the dictionaries.
# Creating a function to extract the info from 'name' key of the dict values in the 'genres' & 'keywords' field
def extract(dic):
    L = []
    for i in ast.literal_eval(dic):
        L.append(i["name"])
    return L
```

```python
# Applying the created function in the 'genres' filed
movies['genres'] = movies['genres'].apply(extract)
movies.head()
```

```python
# Applying the created function in the 'keywords' filed
movies['keywords'] = movies['keywords'].apply(extract)
movies['keywords']
```

```python
# Creating a function to extract top 3 cast info from 'name' key of the dict values in the 'cast' field
def top3extract(word):
    L = []
    increment = 0
    for i in ast.literal_eval(word):
        if increment <3:
            L.append(i['name'])
            increment += 1
        else:
            break
    return L
```

```python
# Applying the created function to 'cast' field
movies['cast'] = movies['cast'].apply(top3extract)
```

```python
# Creating a function to fetch the director name from 'name' key of the dict values in the 'crew' field
def get_director(word):
    L = []
    for i in ast.literal_eval(word):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
# Applying the created function to 'crew' field
movies['crew'] = movies['crew'].apply(get_director)
movies['crew']
```

```python
#Creating a function to remove space between the words to avoid creation of unnecessary tokens of the same words.
def removespace(X):
    L = []
    for i in X:
        L.append(i.replace(' ',''))
    return L    
```

```python
# Removing space between the extracted info from the respective fields of the DF 
movies['genres'] = movies['genres'].apply(removespace)
movies['keywords'] = movies['keywords'].apply(removespace) 
movies['cast'] = movies['cast'].apply(removespace) 
movies['crew'] = movies['crew'].apply(removespace) 
```

## Bag of Words

```python
# Impoting CountVectorizer to convert the text to tokens
from sklearn.feature_extraction.text import CountVectorizer
CV = CountVectorizer(max_features = 10000, stop_words='english')
```

```python
vector = CV.fit_transform(final_movies['context']).toarray()
```

```python
vector.shape
```

```python
vector
```

## Similarity Matrix

```python
# Calculation of cosine_similarity to build the similarity matrix
from sklearn.metrics.pairwise import cosine_similarity
sim_matrix = cosine_similarity(vector)
```

```python
sim_matrix
```

```python
final_movies.head()
```

## Building Recommendation System

```python
final_movies[final_movies['title'] == 'Toy Story'].index[0]
```

```python
# Creating a function to build the recommendation system using the similarity matrix
def recommendation(movie):
    index = final_movies[final_movies['title']== movie].index[0]
    dist = sorted(list(enumerate(sim_matrix[index])),reverse = True, key = lambda x: x[1])
    for i in dist[1:6]:
        print(final_movies.iloc[i[0]].title)
```

## Recommendation System
